```python
dictionary = pd.read_csv('../data/dictionary_cleaned.csv', index_col=0)
```

```python
dictionary
```

```python
char_list = [i for i in dictionary.char]
char_list = set(char_list)
```

```python
chars = {dictionary.char[i]: i for i in range(0,len(dictionary))}
char_dict = {i: {'char': dictionary.char[i],
                 'tone': dictionary.tone_class[i],
                 'rime': dictionary.rime_index[i],
                 'meter': dictionary.meter_class[i]
                 } for i in range(0,len(dictionary))}
```

```python
def character_checker(poem):
    invalid_count = 0
    for i in poem:
        if i not in char_list:
            invalid_count += 1
    if invalid_count == 0:
        return True
    else:
        return False            
```

```python
def text_tokenizer(text):
    return [chars[i] for i in text]
```

```python
def rhyme_checker(tokens):
    if char_dict[tokens[13]]['rime'] == char_dict[tokens[27]]['rime']:
        return True
    else:
        return False
```

```python
def poem_check(poem):
    if character_checker(poem) == True and rhyme_checker(text_tokenizer(poem)) == True:
        return True
    else:
        return False
```

```python
poetry['valid'] = poetry['text'].apply(poem_check)
```

```python
poetry.valid.value_counts()
```

```python
valid_poetry = poetry[poetry['valid']==True]
```

```python
valid_poetry.reset_index(drop=True, inplace=True)
```

```python
valid_poetry['text'].to_csv('../data/final_quatrains.csv', index=0)
```
